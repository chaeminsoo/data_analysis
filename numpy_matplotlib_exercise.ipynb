```python
arr4 = np.array([
    [1,2,3],
    [4,5,6]
], dtype = np.float64)

arr5 = np.array([
    [7,8,9],
    [10,11,12]
], dtype=np.float64)

print('arr4 + arr5 = ')
print(arr4 + arr5,'\n')
print('arr4 - arr5 = ')
print(arr4 - arr5,'\n')
print('arr4 * arr5 = ')
print(arr4 * arr5,'\n')
print('arr4 / arr5 = ')
print(arr4 / arr5,'\n')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
np.random.seed(19920613)

x = np.arange(0.0, 100.0, 5.0)
y = (x*1.5) + np.random.rand(20)*50

plt.scatter(x, y, c='b', alpha=0.5, label = 'scatter point')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='upper left')
plt.title('Scatter plot')
plt.show()
```
