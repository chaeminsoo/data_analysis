```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
source_url = 'https://programmers.co.kr/job?tags=Python&min_salary=&min_career=&min_employees=&order=recent&page=1'
```

```python
# start crawling based on site's HTML structure
req = requests.get(source_url)
html = req.content # original text
soup = BeautifulSoup(html, 'lxml') # (target, method(lxml))
content_table = soup.find(name='ul', attrs={'class':'list-position'})
table_body = content_table.find(name='li', attrs={'class':'list-position-item'})
table_rows = table_body.find_all(name='div', attrs={'class':'item-body'})

page_url_base = 'https://programmers.co.kr/'
page_urls =[]
for index in range(len(table_rows)):
    first_h5 = table_rows[index].find_all('h5')[0]
    h5_url = first_h5.find_all('a')
    if len(h5_url) >0:
        page_url = page_url_base + h5_url[0].get('href')
        page_urls.append(page_url)

page_urls = list(set(page_urls))
for page in page_urls[:5]:
    print(page)
```
