```python
baseUrl = ''
plusUrl = input()
input_page = input("How many pages do you want to crawl: ")
pageNum = 1
url = f'https://programmers.co.kr/job?min_salary=&min_career=&min_employees=&order=recent&page={pageNum}'
# print(plusUrl)
# print(input_page)
# print(url)
```

```python
while pageNum <= input_page:
    url = f'https://programmers.co.kr/job?min_salary=&min_career=&min_employees=&order=recent&page={pageNum}'

    html = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(html, 'lxml')

    title = soup.find_all(class_ = '/job_positions/')
```
