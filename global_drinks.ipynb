```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = 'C:/python-data-analysis-master/python-data-analysis-master/data/drinks.csv'
drinks = pd.read_csv(file_path)
print(drinks.info())
drinks.head(10)
```

```python
drinks.describe()
```

```python
# correlation coefficient between beer_servings and wine_servings
# 'pearson' is one of the most popular methods to figure out correlation coefficient
corr =  drinks[['beer_servings', 'wine_servings']].corr(method='pearson')
print(corr)
```

```python
cols =  ['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']
corr = drinks[cols].corr(method='pearson')
print(corr)
```

```python
import seaborn as sns

cols_view = ['beer', 'spirit', 'wine', 'alcohol']
sns.set(font_scale = 1.5)
hm = sns.heatmap(
    corr.values,
    cbar = True,
    annot = True,
    square = True,
    fmt = '.2f',
    annot_kws={'size':15},
    yticklabels=cols_view,
    xticklabels=cols_view
)

plt.tight_layout()
plt.show()

sns.set(style = 'whitegrid', context = 'notebook')
sns.pairplot(drinks[['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']], height= 2.5)
plt.show()
```
